### Data Sources :


#### Mortagage rate : Freddie Mac, 30-Year Fixed Rate Mortgage Average in the United States [MORTGAGE30US], retrieved from FRED, Federal Reserve Bank of St. Louis; https://fred.stlouisfed.org/series/MORTGAGE30US, September 22, 2023.

#### S&P/Case-Shiller U.S. National Home Price Index  : S&P Dow Jones Indices LLC, S&P/Case-Shiller U.S. National Home Price Index [CSUSHPISA], retrieved from FRED, Federal Reserve Bank of St. Louis; https://fred.stlouisfed.org/series/CSUSHPISA, September 24, 2023.


#### GDP : U.S. Bureau of Economic Analysis  Release: Gross Domestic Product  


#### Producer Price Index (Building Material) : U.S. Bureau of Labor Statistics, Producer Price Index by Industry: Building Material and Supplies Dealers [PCU44414441], retrieved from FRED, Federal Reserve Bank of St. Louis; https://fred.stlouisfed.org/series/PCU44414441, September 22, 2023.

#### Producer Price Index (Transort Warehouse) : U.S. Bureau of Labor Statistics, Producer Price Index by Industry: Transportation and Warehousing Industries [PCUATRNWRATRNWR], retrieved from FRED, Federal Reserve Bank of St. Louis; https://fred.stlouisfed.org/series/PCUATRNWRATRNWR, September 22, 2023.

#### New privately-owned housing units started in the United States: U.S. Census Bureau  Source: U.S. Department of Housing and Urban Development

#### Total housing units in the united states :  ToU.S. Census Bureau, Housing Inventory Estimate: Total Housing Units in the United States [ETOTALUSQ176N], retrieved from FRED, Federal Reserve Bank of St. Louis; https://fred.stlouisfed.org/series/ETOTALUSQ176N, September 22, 2023.

#### Fixed Rate Mortgage : Freddie Mac, 30-Year Fixed Rate Mortgage Average in the United States [MORTGAGE30US], retrieved from FRED, Federal Reserve Bank of St. Louis; https://fred.stlouisfed.org/series/MORTGAGE30US, September 22, 2023.

#### Working age population : Organization for Economic Co-operation and Development, Working Age Population: Aged 15-64: All Persons for United States [LFWA64TTUSM647S], retrieved from FRED, Federal Reserve Bank of St. Louis; https://fred.stlouisfed.org/series/LFWA64TTUSM647S, September 24, 2023.

#### Unemployement rate : U.S. Bureau of Labor Statistics, Unemployment Rate [UNRATE], retrieved from FRED, Federal Reserve Bank of St. Louis; https://fred.stlouisfed.org/series/UNRATE, September 22, 2023.

#### Population Growth : World Bank, Population Growth for the United States [SPPOPGROWUSA], retrieved from FRED, Federal Reserve Bank of St. Louis; https://fred.stlouisfed.org/series/SPPOPGROWUSA, September 24, 2023.

#### Consumer sentiment : University of Michigan, University of Michigan: Consumer Sentiment [UMCSENT], retrieved from FRED, Federal Reserve Bank of St. Louis; https://fred.stlouisfed.org/series/UMCSENT, September 24, 2023.

```python
def interpolate_quarterly_to_monthly(quarterly_data):
    
    quarterly_data.index = pd.to_datetime(quarterly_data.index)
    monthly_data = quarterly_data.resample('M').interpolate(method='linear')
    
    
    monthly_data.reset_index(inplace=True)
    
    return monthly_data
```

```python
#df_target_one[:252].DATE.values

### GDP Transformation

start_date = '2002-01-01'
end_date = '2022-12-01'
date_range = pd.date_range(start=start_date, end=end_date, freq='MS')


df_gdp['DATE'] = pd.to_datetime(df_gdp['DATE'])


df_gdp = pd.DataFrame({'DATE': date_range}).merge(df_gdp, on='DATE', how='left')


df_gdp['DATE'] = df_gdp['DATE'].dt.strftime('%Y-%m-%d')

df_gdp['GDPC1'].interpolate(method='linear', inplace=True)

#df_gdp.to_csv('gdp_data_filled.csv')
```

```python
### Population growth transformation

start_date = '2002-01-01'
end_date = '2022-12-01'


date_range = pd.date_range(start=start_date, end=end_date, freq='MS')


df_pop_growth['DATE'] = pd.to_datetime(df_pop_growth['DATE'])


df_pop_growth.set_index('DATE', inplace=True)


df_pop_growth_resampled = df_pop_growth.resample('MS').asfreq()


df_pop_growth_resampled.reset_index(inplace=True)

df_pop_growth_filled = df_pop_growth_resampled[492:]

#df_pop_growth_filled['DATE'] = df_pop_growth_filled['DATE'].dt.strftime('%Y-%m-%d')

df_pop_growth_filled['DATE'] = pd.to_datetime(df_pop_growth_filled['DATE'])


df_pop_growth_filled['Year'] = df_pop_growth_filled['DATE'].dt.year


df_pop_growth_filled['SPPOPGROWUSA'].fillna(method='ffill', inplace=True)


df_pop_growth_filled.drop(columns=['Year'], inplace=True)

#df_pop_growth_filled.to_csv('pop_growth_filled.csv')
```
